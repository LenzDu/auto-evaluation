# Auto-evaluation Example - Binary Class

```python
%load_ext autoreload
%autoreload 2
import numpy as np
from auto_evaluation import BinaryClassEvaluator
from sklearn.linear_model import LogisticRegression, LinearRegression
```

## Generating Data

```python
sample_size = 100000
np.random.seed(42)

# classification
y = np.random.choice([0, 1], sample_size, p = [0.7, 0.3])
x = y + np.random.randn(sample_size)
x = x.reshape(-1, 1)

clf = LogisticRegression()
clf.fit(x, y)
yfit = clf.predict_proba(x)[:, 1]
```

## Evaluation Report

```python
e = BinaryClassEvaluator(y, yfit)
e.report()
```

## Other Usage

Besides auto generating reports, we can also directly call the plotting functions to draw any specific plots.

```python
e.p_r_curve()
```

```python
e.roc_curve()
```

Use record function to save your result into a local file.

```python
e.record(filename='test_record.txt', note='example', append=False)
with open('test_record.txt') as f:
    for line in f.readlines():
        print(line, end='')
```
